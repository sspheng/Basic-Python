## WHILE LOOPS

1. Python has two primitive loop commands:
        a. While loop

```python
# With the WHILE loop we can execute set of statements till the time it is TRUE

# Print i as long as i is less than 6

i = 0
while i <= 6:
    print (i)
    i = i+1
    
# Remember to incremet i wth 1 otherwise the loop will continue forever. 
```

```python
# With the break statement we can stop the loop even if the while condition is TRUE.

i = 1
while i <6:
    print(i)
    if i==3:
        break
    i+=1
```

```python
# With the CONTINUE statement we can stop the current iteration and move on to the next

i = 0

while i < 6:
    i+=1
    if i==3:
        continue
    print(i)
```

```python
# With the ELSE statement we can run a block of code once when the condition is no longer TRUE

i = 1

while i < 6:
    print(i)
    i+=1
else:
    print("i is greater than 6")
```
