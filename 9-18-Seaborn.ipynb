```python
#polymorphism

#ability of object tot take more than one form 

#method overloading 

class methodoverloading:
    def m1(self):
        print('I am in m1 method')
    def m1(self,a):
        print('i am in seocnd method having 1 para')
    def m1(self,a,b):
        print('I am in third method having 2 para')
    def m1('self,a,b'):
        print('I am in third method having 2 para')
        
        
m=methodoverloading()
m.m1()
```

```python
class pythonmethodoverloadingpossible:
    def m1(self,a=0,b=0,c=0):
        print(a,b,c)
p=pythonmethodoverloadingpossible()
p.m1()
p.m1(2)
p.m1(2,3,4)
```

```python
class TooYoungException(Exception):
    def __init__(self,arg):
        self.msg=arg
class TooOldException(Exception):
    def __init__(self,arg):
        self.msg=arg
```

```python
age=int(input('Enter age'))
if age>65:
    raise TooOldException('You are old to marry')
elif age<18:
    raise TooYoungException('You are too young to marry')
else:
    print('You can marry')
```
